# DBSCAN Model

DBSCAN, which stands for Density-Based Spatial Clustering of Applications with Noise, is a powerful unsupervised machine learning algorithm used for clustering data points in a spatial dataset. Unlike traditional clustering algorithms like K-Means, DBSCAN is particularly well-suited for datasets with irregular shapes and varying densities. It was introduced by Martin Ester, Hans-Peter Kriegel, Jörg Sander, and Xiaowei Xu in 1996.

**Key Concepts:**

1. **Density-Based Clustering:** DBSCAN is a density-based clustering algorithm. It defines clusters as dense regions of data points separated by areas of lower point density. This allows DBSCAN to discover clusters of arbitrary shapes and sizes.

2. **Core Points:** In DBSCAN, a "core point" is a data point that has at least a specified number of data points (MinPts) within a defined distance (Epsilon or ε). Core points are indicative of dense areas within the dataset.

3. **Border Points:** A "border point" is a data point that is within the ε-distance of a core point but does not meet the MinPts criterion. Border points are part of a cluster but are not core points themselves.

4. **Noise Points:** Data points that do not belong to any cluster are considered "noise points." These are often isolated points or points in sparse regions of the dataset.

**Algorithm Steps:**

1. **Initialization:**

   - Choose an ε (Epsilon) and MinPts (minimum number of points required to form a dense region).
   - Select an arbitrary data point not yet visited.

2. **Exploration:**

   - If the selected point is a core point, create a new cluster and assign it to the cluster. 
   - Expand the cluster by adding all directly reachable core points to it.

3. **Expansion:**

   - Continue expanding the cluster until no more core points can be added.
   - If a border point is reached during expansion, mark it as a border point and continue.

4. **Termination:**

   - Repeat the process for unvisited data points until all data points have been processed.

5. **Result:**

   - The final result of DBSCAN is a set of clusters and noise points.

**Advantages:**

- DBSCAN can find clusters of arbitrary shapes and is robust to outliers.
- It does not require specifying the number of clusters in advance, making it suitable for datasets with varying cluster densities.

**Challenges:**

- Tuning the ε and MinPts parameters can be challenging and may require domain knowledge.
- DBSCAN is sensitive to the choice of distance metric.

**Applications:**

DBSCAN has been widely used in various fields, including image processing, geospatial analysis, anomaly detection, and customer segmentation.

**Conclusion:**

In summary, DBSCAN is a versatile clustering algorithm that excels in discovering clusters of different shapes and sizes, making it a valuable tool in unsupervised machine learning. Its ability to identify noise points and outliers provides added practicality in real-world applications.

```python
# Scatter plot of the 2 first principal components

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

label_color_mapping = {
     0: 'blue',
     1: 'red',
     2: 'green',
     3: 'purple',
     4: 'orange',
     5: 'brown',
    -1: 'gray'  # For the noise points
}

label_color_mapping_cats = {
     'build_wind_float' : 'blue',
     'vehic_wind_float': 'red',
     'tableware': 'green',
     'build_wind_non-float': 'purple',
     'headlamps': 'orange',
     'containers': 'brown',
}

d_labels = {0: 'build_wind_float',
                            1: 'vehic_wind_float',
                            2: 'tableware',
                            3: 'build_wind_non-float',
                            4: 'headlamps',
                            5: 'containers'
                            }
def plot_clusters(tf, d_labels, d_labels_color, d_labels_color_cats):


    pca = PCA(n_components=2)
    X = tf.data_loader.df.drop(columns="Type")
    y = tf.trained_model.labels_
    X = pca.fit_transform(X)
    y_true = tf.data_loader.df["Type"]
    y_true = y_true.replace(d_labels)
    y_wo_noise = y[y != -1]
    y_true_wo_noise = y_true[y != -1]
    X_wo_noise = X[y != -1]
    X_noise = X[y == -1]


    fig, ax = plt.subplots(3, 2, figsize=(15,15))
    sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, palette=d_labels_color, ax=ax[0, 0])
    sns.scatterplot(x=X[:,0], y=X[:,1], hue=y_true, palette=d_labels_color_cats, ax=ax[0, 1])
    sns.scatterplot(x=X_wo_noise[:,0], y=X_wo_noise[:,1], hue=y_wo_noise, palette=d_labels_color, ax=ax[1, 0])
    sns.scatterplot(x=X_wo_noise[:,0], y=X_wo_noise[:,1], hue=y_true_wo_noise, palette=d_labels_color_cats, ax=ax[1, 1])
    sns.scatterplot(x=X_noise[:,0], y=X_noise[:,1], hue=y[y==-1], palette=d_labels_color, ax=ax[2, 0])
    sns.scatterplot(x=X_noise[:,0], y=X_noise[:,1], hue=y_true[y==-1], palette=d_labels_color_cats, ax=ax[2, 1])

    # Get explained variance ratio 2 first components
    exp_var = pca.explained_variance_ratio_.sum()
    clustered_ratio = len(y_wo_noise) / len(y)
    fig.suptitle(f"{tf.training_params}\nClustered: {clustered_ratio:.2%}\nPCA: {exp_var:.2%}")
    ax[0, 0].set_title("Predicted labels (with noise)")
    ax[0, 1].set_title("True labels (with noise)")
    ax[1, 0].set_title("Predicted labels (without noise)")
    ax[1, 1].set_title("True labels (without noise)")
    ax[2, 0].set_title("Predicted labels (only noise)")
    ax[2, 1].set_title("True labels (only noise)")

    # y labels are PCA 2 for all subplots
    for i in range(3):
        for j in range(2):
            ax[i, j].set_xlabel("PCA 1")
            ax[i, j].set_ylabel("PCA 2")

    fig.tight_layout()
    plt.show()

    count_noise = tf.data_loader.df[y == -1].Type.value_counts() / tf.data_loader.df.Type.value_counts()
    count_noise = count_noise.map(lambda x: f"{x:.2%}").reset_index().rename(columns={"index": "Type", "Type": "Noise ratio"})
    count_noise['Type'] = count_noise['Type'].replace(d_labels)
    count_noise["Ratio_from_total"] = tf.data_loader.df[y == -1].Type.value_counts(normalize=True).values
    count_noise["Ratio_from_total"] = count_noise["Ratio_from_total"].map(lambda x: f"{x:.2%}")
    print(count_noise)

plot_clusters(tf, d_labels, 
              label_color_mapping, label_color_mapping_cats)
```

```python
d_hyperparams = {'eps': 0.1,
                'min_samples': 3,
                'metric': 'cosine',
                'algorithm': 'auto'}
tf = TrainFlow(ds_name, model, hyperparams=d_hyperparams, target=target, metrics=ls_metrics)
tf.run()
plot_clusters(tf, d_labels,
              label_color_mapping, label_color_mapping_cats)
```
