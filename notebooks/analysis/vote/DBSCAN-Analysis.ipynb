```python
import pandas as pd
from src.trainflow.TrainFlow import TrainFlow
from src.clustering.DBSCAN import DBSCAN
import matplotlib.pyplot as plt
import os


ds_name = "vote"
model = DBSCAN
target = "Class"
ls_metrics = ['adjusted_rand_score', 'v_measure_score', 
                'silhouette_score', 'calinski_harabasz_score']

d_labels = {
            0: 'democrat',
            1: 'republican',
            }

tf = TrainFlow(ds_name, model, target=target, metrics=ls_metrics, d_plot_params=d_labels)
tf.run()
```
